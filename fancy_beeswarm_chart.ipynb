```python
# Function to clean the title column using regex
def clean_title(title):
    # Standardize by removing comma followed by spaces and numbers, and everything after it
    return re.sub(r',\s*\d.*', '', title)

# Apply the 'clean_title' function to the 'titles' column
df['titles'] = df['titles'].apply(clean_title)
```

```python
# Fancy!
df['titles'].head(10)
```

```python
# Get the path to the images folder
images_folder = "/Users/juditecypreste/project 4/catch_images/images"
```

```python
# Function to extract the filename from a URL
def get_file_name(url):
    return url.split('/')[-1]
```

```python
# List to store the new names
new_names = []
```

```python
# Iterate through the DataFrame and rename the images
for index, row in df.iterrows():
    url = row['images']
    title = row['titles']
    current_file_name = get_file_name(url)
    current_path = os.path.join(images_folder, current_file_name)
    new_file_name = title + '.jpg'  # Assuming all images are in JPG format, you can adjust if needed
    new_path = os.path.join(images_folder, new_file_name)

    # Check if the file exists before renaming
    if os.path.exists(current_path):
        os.rename(current_path, new_path)
        print(f"File renamed: {current_file_name} -> {new_file_name}")
        new_names.append(new_file_name)
    else:
        print(f"File not found: {current_file_name}")
        new_names.append(current_file_name)
```

```python
# Create a new DataFrame with the same information and the new 'address' column
data_paintings = df.copy()
data_paintings['address'] = new_names

data_paintings.head()
```
